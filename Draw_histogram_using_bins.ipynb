```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
n=1000
k =40
data = np.exp(np.random.randn(n)/2)
```

```python
plt.hist(data,k)
plt.show()
```

```python
r = 2*stats.iqr(data)*n**(-1/3)
b = np.ceil((max(data)-min(data))/r)

plt.hist(data,int(b))
plt.show()
```
